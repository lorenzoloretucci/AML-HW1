```python
def gaussderiv(img, sigma):

    smooth_img = gauss_module.gaussianfilter(img, sigma)

    imgDx = np.diff(smooth_img, axis = 1)
    imgDy = np.diff(smooth_img, axis = 0)
    
    imgDx = np.column_stack((imgDx, np.zeros(imgDx.shape[0])))
    imgDy = np.vstack ((imgDy, np.zeros(imgDy.shape[1])) )
    
    return imgDx, imgDy
```

```python
img_c = np.array(Image.open('graf.png')).astype('double')
img = rgb2gray(img_c)
[imgDx, imgDy] = gaussderiv(img, 7.0)

plt.figure(figsize=(20,10))
ax1 = plt.subplot(1,3,1)
ax2 = plt.subplot(1,3,2)
ax3 = plt.subplot(1,3,3)
plt.sca(ax1)
plt.imshow(imgDx, cmap='gray')  ##Derivative filter x direction
plt.sca(ax2)
plt.imshow(imgDy, cmap='gray')  ##Derivative filter y direction
plt.sca(ax3)
imgmag = np.sqrt(imgDx**2 + imgDy**2)  ##Magnitude
plt.imshow(imgmag, cmap='gray')
plt.show()
```

```python
def convolve2d(image, kernel):
    """
    This function which takes an image and a kernel and returns the convolution of them.
    :param image: a numpy array of size [image_height, image_width].
    :param kernel: a numpy array of size [kernel_height, kernel_width].
    :return: a numpy array of size [image_height, image_width] (convolution output).
    """
    kernel = kernel.reshape(kernel.size, 1)
    # Flip the kernel
    #kernel = np.flipud(np.fliplr(kernel))
    # convolution output
    output = np.zeros_like(image)
    
    # Add zero padding to the input image
    image_padded = np.zeros((image.shape[0] + kernel.size, image.shape[1] + kernel.size))
    image_padded[1:-kernel.size+1, 1:-kernel.size+1] = image

    # Loop over every pixel of the image
    for y in range(image.shape[0]):
        #for y in range(image.shape[0]):
            # element-wise multiplication of the kernel and the image
            #output[y, x] = (kernel * image_padded[y: y+kernel.shape[0], x: x+kernel.shape[1]]).sum()
        output[y] = (kernel * image_padded[y: y+kernel.size]).sum()
    for x in range(image.shape[1]):
        output[:,x] = (kernel.T * image_padded[:, x: x+kernel.size]).sum()

    return output
```
